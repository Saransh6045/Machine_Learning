Method to Handle Missing Values:
1. Imputation
2. Dropping

```python
#Importing the libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loadin gthe dataset to a dataframe
dataset = pd.read_csv('/content/Placement_Dataset.csv')
```

```python
dataset.head()
#here NaN represents missing values
```

```python
dataset.shape
```

```python
#finding the missing values in the dataset
dataset.isnull().sum()

```

Central Tendencies:
1. Mean
2. Median
3. Mode

```python
#analyse the distribution of datan in the salary
fig, ax = plt.subplots(figsize=(8,8))
sns.distplot(dataset.salary)
plt.show()
```

Replace the missing values with median

```python
dataset['salary'].fillna(dataset['salary'].median(), inplace=True)
```

```python
dataset.isnull().sum()
```

Replacing the missing values with mean value

```python
dataset['salary'].fillna(dataset['salary'].mean(), inplace=True)
```

Replace the missing values with mode

```python
dataset['salary'].fillna(dataset['salary'].mode(), inplace=True)
```

Dropping

```python
salary_dataset = pd.read_csv('/content/Placement_Dataset.csv')
```

```python
salary_dataset.shape
```

```python
#dropping the missing values
salary_dataset.dropna(how='any')
```

#Label Encoding


*   converting the labels into numeric form


```python
#importing the librtaries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

Label Encoding of Breast Cancer Dataset

```python
#importing the data from csv file to pandas dataframe
bc_dataset = pd.read_csv('/content/breast_cancer_data.csv')
bc_dataset.head()
bc_dataset.tail()
```

```python
#in this dataset, we have to apply label encoding dataset on diagnosis column only. It contains two categories: M & B

#finding the count of different labels
bc_dataset['diagnosis'].value_counts()
```

```python
#Loading the label encoder function
label_encode = LabelEncoder()
```

```python
labels = label_encode.fit_transform(bc_dataset.diagnosis)
```

```python
#appending the new diagnosis in the dataset
bc_dataset['diagnosis_value'] = labels

bc_dataset.head()
```

*  0 represents Benign
*  1 represents Malignant



```python
bc_dataset['diagnosis_value'].value_counts()
```

```python
#The assignation of 0 and 1 depends upon the alphabetical order of the words. Here B comes before M, so it is assigned 0.
```
